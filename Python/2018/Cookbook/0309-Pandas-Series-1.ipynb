```python
# What hapens when we have Nan? Does count, mean and sum work? How?

y={0:1.5,1:2,2:3,3:np.nan,4:4,5:5}
s1=pd.Series(y)
s1
```

```python
s1.isnull()
```

```python
print (s1.isnull() ) #isna
print ( '----------')
#print (s1.notnull()) 
#print ('----------')
print ('count = ', s1.count() )
print ('sum = ', s1.sum())
print ('mean= ', s1.mean())
print ('median= ', s1.median())
print ('our mean =' ,s1.sum()/(len(s1) ))# be careful subtract out all of the na's otherwise your dividing too much
print ('----------------')
print ('the describe method shows that nans don\'t count in arithmetic \n ', s1.describe())
print(' value counts  ', s1.value_counts())
```
